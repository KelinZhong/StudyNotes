# Business decision on implementing ML models — overview
**Topics:** Decision Pipeline · Folder Map · When to Use Which Method · Quick Reference

## What this folder is for

- This folder answers one question: **given a model result, should we act on it?**
- It does not explain how models work or how to build them — those live in the other folders
- The focus is on judgment: reading results critically, translating them into business value, and making a defensible ship or no-ship decision

### What this folder is NOT
- Not a modeling tutorial — see `Machine Learning/` and `Deep Learning/`
- Not an experiment design guide — see `Experiment/`
- Not a causal inference implementation guide — see `Causal ML/`
- Not a deployment implementation guide — see `Data Science Workflow/ds7_deployment.ipynb`

---
## The decision pipeline

Every model implementation decision follows the same pipeline:

```
1. Problem framing & metric selection
         ↓
2. Offline evaluation — does the model work?
         ↓
3. A/B test or causal model — does it work in the real world?
         ↓
4. Cost-benefit — is it worth it?
         ↓
5. Ship / no-ship decision
         ↓
6. Post-launch review
```

### Key principle at each stage

| Stage | Key principle |
|---|---|
| Problem framing | Define success criteria before seeing any results |
| Offline evaluation | Statistical significance is necessary but not sufficient |
| A/B / causal | Check Sample Ratio Mismatch (SRM) first; pre-registration determines whether the result is confirmatory or exploratory |
| Cost-benefit | Use the lower bound of the 95% Confidence Interval (CI), not the point estimate |
| Ship decision | Reversibility determines the evidence bar — irreversible failures require stronger evidence |
| Post-launch | Set alert thresholds before launch, not after something goes wrong |

## When to use which evaluation method

### Choosing between A/B and causal models

| Situation | Recommended method | Reason |
|---|---|---|
| Can randomize, cost is acceptable | A/B test | Gold standard — no assumptions needed beyond randomization |
| Cannot randomize, have pre/post data and a control group | Difference-in-Differences (DiD) | Exploits time structure; requires parallel trends assumption |
| Cannot randomize, have an exogenous instrument | Instrumental Variables (IV) | Isolates causal effect via instrument; requires strong first stage |
| Treatment assigned by a threshold rule | Regression Discontinuity Design (RDD) | Near-random assignment at cutoff; local estimate only |
| Have rich observational data, no instrument or threshold | Propensity Score Matching (PSM) | Last resort; most sensitive to unmeasured confounding |
| Want to personalize treatment, have A/B data | Uplift model | Estimates individual-level causal effect; requires unconfoundedness |

### Strength of evidence hierarchy

| Rank | Method | Evidence strength |
|---|---|---|
| 1 | Well-powered, pre-registered A/B test | Strongest — causal by design |
| 2 | RDD with clean density and covariate balance | Strong — near-random at threshold |
| 3 | IV with strong first stage and credible exclusion | Strong — isolates exogenous variation |
| 4 | DiD with flat pre-trend and multiple pre-periods | Moderate — depends on parallel trends |
| 5 | Uplift model with A/B validation of segment ranking | Moderate — depends on unconfoundedness |
| 6 | PSM with good overlap and high Rosenbaum bounds | Weakest — most sensitive to hidden confounders |

## Quick reference — common decision mistakes

| Mistake | Consequence | Correct practice |
|---|---|---|
| Checking significance without checking SRM | A/B result is invalid but looks valid | Always run SRM check before reading any metric |
| Using point estimate of lift in business case | Overestimates expected value | Use lower bound of 95% CI |
| Choosing primary metric after seeing results | Inflated false positive rate | Pre-register primary metric before launch |
| Treating exploratory causal result as confirmatory | Overconfident ship decision | Require A/B validation or stronger evidence bar |
| Ignoring opportunity cost | Suboptimal team allocation | Compare model ROI against next best alternative |
| No rollback plan before shipping | Slow incident response | Define rollback procedure and auto-rollback triggers before deployment |
| Skipping slice analysis | Global gain hides segment harm | Always check performance across key segments before shipping |
| Writing model card after an incident | Governance failure | Complete model card before shipping, not after |
