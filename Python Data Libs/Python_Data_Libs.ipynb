## How to Use These Notes

- **Reference notes** (`Reference_*.ipynb`) are **syntax-first** — look up how to call a function, what arguments it takes, and what the common mistakes are. Use the *When to Use* table at the top of each file to navigate quickly.

- **Data manipulation notes** (`DM_*.ipynb`) are **pattern-first** — each section answers "you have data that looks like X, here is the full pipeline to solve it." They assume you already know the Pandas/NumPy API.

- Every section ends with a **Common mistakes** block and every file ends with a **Decision Guide** — use these during interviews to sanity-check your approach.
