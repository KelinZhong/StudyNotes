## §1 — List

### §1a — Creation & copying

```python
# Creation
lst = []                          # empty list
lst = [0] * n                     # list of n zeros
lst = list(range(n))              # [0, 1, 2, ..., n-1]
lst = [i**2 for i in range(n)]    # list comprehension
lst = [0] * m                     # 1D
grid = [[0] * n for _ in range(m)]  # 2D — correct
# grid = [[0] * n] * m  ← WRONG: all rows share the same reference

# Copying
import copy
lst2 = lst[:]                     # shallow copy — new list, same object references
lst2 = lst.copy()                 # shallow copy — identical to lst[:]
lst2 = list(lst)                  # shallow copy
lst2 = copy.deepcopy(lst)         # deep copy — recursively copies nested objects

# Shallow vs deep copy
# Shallow: safe for flat lists of primitives (ints, strings)
# Deep:    required for nested lists or lists of mutable objects
```

### §1b — Modification

```python
lst = [1, 2, 3]

# Add
lst.append(4)           # [1, 2, 3, 4] — O(1) amortized
lst.extend([5, 6])      # [1, 2, 3, 4, 5, 6] — add multiple
lst.insert(1, 99)       # [1, 99, 2, ...] — O(n) — inserts at index 1

# Remove
lst.pop()               # removes and returns LAST element — O(1)
lst.pop(0)              # removes and returns element at index 0 — O(n)
lst.remove(99)          # removes FIRST occurrence of value 99 — O(n), raises ValueError if not found
del lst[1]              # delete by index — O(n)
del lst[1:3]            # delete slice
lst.clear()             # empty the list

# Swap — in-place, no temp variable needed
lst[i], lst[j] = lst[j], lst[i]
```

### §1c — Search

```python
lst = [1, 2, 3, 2, 1]

3 in lst                # True — O(n)
lst.index(2)            # 1 — first occurrence, raises ValueError if not found
lst.count(2)            # 2 — count occurrences
```

### §1d — Sorting

```python
import functools

lst = [3, 1, 2]

# sort() — in-place, returns None
lst.sort()                          # ascending
lst.sort(reverse=True)              # descending
lst.sort(key=lambda x: x[1])        # sort by second element
lst.sort(key=lambda x: (x[1], x[0])) # sort by multiple keys
lst.sort(key=abs)                   # sort by absolute value

# sorted() — returns NEW sorted list, original unchanged
new_lst = sorted(lst)               # ascending
new_lst = sorted(lst, reverse=True) # descending
new_lst = sorted(lst, key=len)      # sort strings by length

# Custom comparator — when key= is not enough
# Example: sort so that concatenated result is largest (LC 179)
def compare(a, b):
    if a + b > b + a: return -1     # a should come first
    elif a + b < b + a: return 1
    else: return 0

nums = ['3', '30', '34']
nums.sort(key=functools.cmp_to_key(compare))

# sort() vs sorted()
# sort()   → in-place, modifies original, returns None
# sorted() → returns new list, original unchanged
```

### §1e — Iteration idioms

```python
lst = ['a', 'b', 'c']

# enumerate — index + value
for i, val in enumerate(lst):       # i=0,1,2; val='a','b','c'
    pass
for i, val in enumerate(lst, 1):    # start index at 1
    pass

# zip — pair elements from multiple iterables
for a, b in zip(lst1, lst2):        # stops at shortest
    pass
list(zip([1,2,3], ['a','b','c']))   # [(1,'a'), (2,'b'), (3,'c')]

# zip to unzip (transpose)
matrix = [[1,2],[3,4],[5,6]]
transposed = list(zip(*matrix))     # [(1,3,5), (2,4,6)]

# List comprehension
[x for x in lst if x > 0]          # filter
[x**2 for x in lst]                # transform
[x for row in matrix for x in row] # flatten 2D list
[[0]*n for _ in range(m)]          # 2D grid initialization

# any / all
any(x > 0 for x in lst)            # True if at least one element > 0
all(x > 0 for x in lst)            # True if all elements > 0
sum(x > 0 for x in lst)            # count of elements > 0 (True == 1)
```

**Common mistakes:**
- `[[0]*n]*m` creates m references to the same row — use `[[0]*n for _ in range(m)]`
- `lst.sort()` returns `None` — never do `lst = lst.sort()`; use `sorted(lst)` instead
- `list.pop(0)` is O(n) — use `deque.popleft()` for O(1) front removal
- `list.insert(0, val)` is O(n) — use `deque.appendleft()` for O(1) front insertion
- `list.remove(val)` raises `ValueError` if val not in list — check `val in lst` first

## §2 — Dictionary

### §2a — Creation & access

```python
from collections import defaultdict, Counter, OrderedDict

# Creation
d = {}                              # empty dict
d = {'a': 1, 'b': 2}
d = dict(a=1, b=2)
d = dict(zip(keys, values))         # from two lists
d = {k: v for k, v in items}       # dict comprehension

# Access
d['a']                              # KeyError if not found
d.get('a')                          # None if not found — safe
d.get('a', 0)                       # 0 if not found — with default

# Modification
d['c'] = 3                          # add or update
del d['a']                          # KeyError if not found
d.pop('a')                          # removes and returns — KeyError if not found
d.pop('a', None)                    # removes and returns None if not found — safe
d.update({'c': 3, 'd': 4})          # merge another dict
```

### §2b — Iteration

```python
d = {'a': 1, 'b': 2, 'c': 3}

d.keys()                            # dict_keys(['a', 'b', 'c'])
d.values()                          # dict_values([1, 2, 3])
d.items()                           # dict_items([('a',1), ('b',2), ('c',3)])

for k in d:                         # iterate over keys
    pass
for k, v in d.items():              # iterate over key-value pairs
    pass

# Check existence
'a' in d                            # True — O(1) key lookup
'a' not in d
```

### §2c — defaultdict & setdefault

```python
# defaultdict — auto-initializes missing keys
d = defaultdict(int)                # missing keys default to 0
d['a'] += 1                         # no KeyError

d = defaultdict(list)               # missing keys default to []
d['key'].append(1)                  # no KeyError

d = defaultdict(set)                # missing keys default to set()
d = defaultdict(lambda: float('inf'))  # custom default

# setdefault — initialize only if key is missing, then return value
d.setdefault('key', []).append(1)   # equivalent to defaultdict(list) pattern
d.setdefault('count', 0)            # initialize to 0 if not present

# defaultdict vs setdefault
# defaultdict: set once at creation, applies to ALL missing keys
# setdefault:  set per key, more flexible, works on regular dict
```

### §2d — Counter

```python
from collections import Counter

c = Counter('aabbc')                # Counter({'a':2, 'b':2, 'c':1})
c = Counter([1, 2, 2, 3])           # works on any iterable

c['a']                              # 2 — returns 0 for missing keys (not KeyError)
c.most_common(2)                    # [('a',2), ('b',2)] — top 2 most frequent
c.most_common()[:-3:-1]             # bottom 2 least frequent

# Counter arithmetic
c1 + c2                             # combine counts
c1 - c2                             # subtract (keeps only positive)
c1 & c2                             # intersection (min of each)
c1 | c2                             # union (max of each)

# Check if one string's chars are subset of another
all(Counter(s1)[ch] <= Counter(s2)[ch] for ch in Counter(s1))
# Or: not (Counter(s1) - Counter(s2))  — empty counter is falsy
```

### §2e — OrderedDict (LRU Cache pattern)

```python
from collections import OrderedDict

# OrderedDict maintains insertion order and supports move_to_end
# Used for LRU Cache implementation (LC 146)
cache = OrderedDict()
cache[key] = val                    # insert
cache.move_to_end(key)              # move to most recently used (end)
cache.move_to_end(key, last=False)  # move to front (least recently used)
cache.popitem(last=True)            # remove and return last item (MRU)
cache.popitem(last=False)           # remove and return first item (LRU)

# Note: Python 3.7+ regular dict maintains insertion order too
# Use OrderedDict when you need move_to_end or popitem(last=)
```

**Common mistakes:**
- `d[key]` raises `KeyError` — always use `d.get(key, default)` when key may be missing
- `Counter` returns 0 for missing keys — unlike regular dict which raises `KeyError`
- `Counter(s1) - Counter(s2)` drops negative counts — use for subset checking
- Iterating over dict while modifying it — make a copy of keys first: `for k in list(d.keys())`

Problems: LC 1, LC 49, LC 146, LC 347, LC 560

## §3 — Set

### §3a — Creation & modification

```python
s = set()                           # empty set — NOT {} (that's a dict)
s = {1, 2, 3}
s = set([1, 2, 2, 3])               # {1, 2, 3} — deduplicates

# Modification
s.add(4)                            # add element — O(1)
s.remove(4)                         # remove — raises KeyError if not found
s.discard(4)                        # remove — NO error if not found (safer)
s.pop()                             # remove and return arbitrary element
s.clear()                           # empty the set

# Lookup
4 in s                              # O(1) — main advantage over list
```

### §3b — Set operations

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Operations — return new set
a | b                               # {1,2,3,4,5,6} — union
a & b                               # {3,4}          — intersection
a - b                               # {1,2}          — difference (in a, not in b)
a ^ b                               # {1,2,5,6}      — symmetric difference

# Method equivalents
a.union(b)
a.intersection(b)
a.difference(b)
a.symmetric_difference(b)

# Subset / superset checks
a.issubset(b)                       # True if all elements of a are in b
a.issuperset(b)                     # True if a contains all elements of b
a.isdisjoint(b)                     # True if a and b have no common elements
a <= b                              # subset
a >= b                              # superset

# frozenset — immutable set, usable as dict key or set element
fs = frozenset([1, 2, 3])
d = {frozenset([1,2]): 'pair'}      # use as dict key
s = {frozenset([1,2]), frozenset([3,4])}  # set of sets
```

**Common mistakes:**
- `set()` not `{}` for empty set — `{}` creates an empty dict
- `remove()` vs `discard()` — use `discard()` to avoid `KeyError` when element may not exist
- Sets are unordered — cannot index or slice a set
- `frozenset` required when using a set as a dict key or putting it inside another set

Problems: LC 128, LC 217, LC 349, LC 202

## §4 — Tuple

### §4a — Creation & immutability

```python
t = (1, 2, 3)                       # tuple
t = (1,)                            # single-element tuple — comma required
t = tuple([1, 2, 3])                # from list
t = tuple('abc')                    # ('a', 'b', 'c')

# Immutable — cannot change elements
# t[0] = 5  → TypeError

# Can be used as dict key or set element (list cannot)
d = {(0, 1): 'origin'}             # tuple as dict key
visited = {(i, j) for i, j in ...} # set of coordinate tuples

# Unpacking
a, b = (1, 2)
a, b, c = (1, 2, 3)
first, *rest = (1, 2, 3, 4)        # first=1, rest=[2,3,4]
*init, last = (1, 2, 3, 4)         # init=[1,2,3], last=4
a, b = b, a                         # swap using tuple unpacking
```

### §4b — namedtuple

```python
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])
p = Point(1, 2)
p.x                                 # 1 — access by name
p[0]                                # 1 — access by index
p._asdict()                         # {'x': 1, 'y': 2}

# Common use: graph edges, intervals, coordinates with named fields
Interval = namedtuple('Interval', ['start', 'end'])
Edge = namedtuple('Edge', ['weight', 'node'])  # for heap — weight first for comparison
```

**Common mistakes:**
- Forgetting comma for single-element tuple — `(1)` is just `1`; use `(1,)`

- Tuple in heap: comparison is element by element — put the priority field first

## §5 — deque (`collections.deque`)

### §5a — Operations & complexity

```python
from collections import deque

dq = deque()                        # empty deque
dq = deque([1, 2, 3])               # from iterable
dq = deque(maxlen=k)                # fixed-size — auto-evicts oldest when full

# O(1) operations on both ends
dq.append(4)                        # add to right
dq.appendleft(0)                    # add to left
dq.pop()                            # remove from right
dq.popleft()                        # remove from left — O(1) vs list O(n)

# Access
dq[0]                               # left end — O(1)
dq[-1]                              # right end — O(1)
dq[k]                               # arbitrary index — O(n) for deque

# Rotation
dq.rotate(1)                        # rotate right by 1
dq.rotate(-1)                       # rotate left by 1

# deque vs list — when to use which
# deque:  O(1) appendleft / popleft — use for BFS queue, sliding window
# list:   O(1) append / pop at end  — use for stack, DFS
```

### §5b — BFS pattern

```python
from collections import deque

# Always use deque for BFS — never list
# list.pop(0) is O(n); deque.popleft() is O(1)
queue = deque([start])
visited = {start}
while queue:
    node = queue.popleft()          # O(1)
    for neighbor in graph[node]:
        if neighbor not in visited:
            visited.add(neighbor)
            queue.append(neighbor)

# Fixed-size sliding window with maxlen
window = deque(maxlen=k)
for val in nums:
    window.append(val)              # auto-evicts oldest when len > k
    if len(window) == k:
        process(window)
```

- `deque(maxlen=k)` silently drops oldest elements — check length before processing

## §6 — int & float

### §6a — Key properties

```python
# Python integers have no overflow — arbitrary precision
# Unlike Java/C++ where int overflows at 2^31 - 1
2**100                              # works fine

# Infinity
float('inf')                        # positive infinity
float('-inf')                       # negative infinity
float('inf') > 10**18               # True — use as initial min/max value

# Division
7 / 2                               # 3.5  — true division (always float)
7 // 2                              # 3    — floor division (rounds toward -inf)
7 % 2                               # 1    — modulo
-7 // 2                             # -4   — floors toward -inf (not toward 0)
-7 % 2                              # 1    — Python modulo always non-negative for positive divisor
int(-7 / 2)                         # -3   — truncation toward zero (C-style)

# Absolute value
abs(-5)                             # 5

# Power
2**10                               # 1024
pow(2, 10)                          # 1024
pow(2, 10, 1000)                    # 24  — modular exponentiation O(log n), no overflow
```

### §6b — Math functions

```python
import math

math.ceil(3.2)                      # 4  — round up
math.floor(3.8)                     # 3  — round down
math.sqrt(16)                       # 4.0
math.log(8, 2)                      # 3.0 — log base 2
math.log2(8)                        # 3.0
math.log10(100)                     # 2.0
math.gcd(12, 8)                     # 4   — greatest common divisor
math.lcm(4, 6)                      # 12  — least common multiple (Python 3.9+)
math.gcd(a, b, c)                   # GCD of multiple numbers (Python 3.9+)
math.isqrt(10)                      # 3   — integer square root (floor)
math.inf                            # same as float('inf')

# Integer ceiling division without import
(a + b - 1) // b                    # equivalent to math.ceil(a / b)
-(-a // b)                          # also ceiling division
```

**Common mistakes:**
- `//` floors toward negative infinity — `-7 // 2 == -4` not `-3`; use `int(a/b)` for C-style truncation
- `math.sqrt` returns float — use `math.isqrt` for exact integer square root
- `pow(x, n, mod)` is much faster than `(x**n) % mod` for large n — always use the 3-argument form
- No overflow in Python — but intermediate values in `(x**n) % mod` without `pow()` can be astronomically large

Problems: LC 29, LC 50, LC 204, LC 1492

## §7 — bool & None

### §7a — bool

```python
# bool is a subclass of int in Python
True == 1                           # True
False == 0                          # True
True + True                         # 2

# Count elements satisfying a condition
sum(x > 0 for x in nums)           # count positive numbers
sum(c.isupper() for c in s)        # count uppercase letters
sum(1 for x in nums if x > 0)      # explicit form

# Falsy values in Python
# False, 0, 0.0, '', [], {}, set(), None, deque()
if not lst: ...                     # True when list is empty
if lst: ...                         # True when list is non-empty

# Boolean operators
# and: returns first falsy value, or last value if all truthy
# or:  returns first truthy value, or last value if all falsy
x = a or b                          # x = a if a is truthy, else b
x = a and b                         # x = b if a is truthy, else a
```

### §7b — None

```python
x = None

x is None                           # True — correct way to check
x is not None
x == None                           # works but not recommended — use 'is'

# None as default sentinel
def func(val=None):
    if val is None:
        val = []

# TreeNode / ListNode null check
if not node: return                  # True when node is None
if node is None: return              # explicit form
```

**Common mistakes:**
- `== None` instead of `is None` — functionally equivalent for None but `is` expresses intent
- Using mutable default argument `def func(lst=[])` — shared across all calls; use `None` instead
- `sum(condition for x in lst)` relies on `True == 1` — works but make intent clear with a comment

Problems: LC 104, LC 206 (node is None check)

## §8 — Built-in Functions

```python
# min / max
min(a, b)                           # smaller of two values
min(lst)                            # smallest in list
min(lst, key=len)                   # smallest by length
min(lst, default=0)                 # 0 if lst is empty
max(a, b, c)                        # largest of multiple values

# sum
sum(lst)                            # sum of list
sum(lst, start)                     # sum + start value

# len
len(lst)                            # O(1) for list, dict, set, string, deque

# range
range(n)                            # 0 to n-1
range(a, b)                         # a to b-1
range(a, b, step)                   # with step
range(n-1, -1, -1)                  # n-1 down to 0 (reverse)

# map / filter
list(map(int, ['1','2','3']))       # [1, 2, 3] — apply function to each element
list(filter(lambda x: x>0, lst))   # keep only positive elements

# sorted with key
sorted(lst, key=lambda x: -x)      # descending
sorted(words, key=lambda w: (len(w), w))  # by length then alphabetically

# enumerate
list(enumerate(['a','b']))          # [(0,'a'), (1,'b')]

# zip
list(zip([1,2], ['a','b']))         # [(1,'a'), (2,'b')]

# isinstance
isinstance(x, int)                  # True if x is int
isinstance(x, (int, float))         # True if x is int or float

# round
round(3.14159, 2)                   # 3.14
round(3.5)                          # 4 (banker's rounding — rounds to even)

# divmod
divmod(10, 3)                       # (3, 1) — quotient and remainder at once
```

**Common mistakes:**
- `min([])` raises `ValueError` — use `min(lst, default=0)` for empty lists
- `map()` returns an iterator — wrap in `list()` if you need to index or reuse
- `round(0.5) == 0` in Python 3 — banker's rounding rounds to even; use `math.ceil/floor` for predictable behavior

Problems: LC 1, LC 15, LC 56

## §9 — `itertools` & `functools`

```python
import itertools
import functools

# itertools — combinatorics
list(itertools.permutations([1,2,3]))       # all permutations
list(itertools.permutations([1,2,3], 2))    # permutations of length 2
list(itertools.combinations([1,2,3], 2))    # [(1,2),(1,3),(2,3)] — no repeats
list(itertools.combinations_with_replacement([1,2,3], 2))  # repeats allowed
list(itertools.product([1,2], [3,4]))       # cartesian product [(1,3),(1,4),(2,3),(2,4)]
list(itertools.product([0,1], repeat=3))    # all binary strings of length 3

# itertools — iteration
list(itertools.chain([1,2], [3,4]))         # [1,2,3,4] — flatten one level
list(itertools.chain.from_iterable([[1,2],[3,4]]))  # same
list(itertools.accumulate([1,2,3,4]))       # [1,3,6,10] — running sum
list(itertools.accumulate([1,2,3], max))    # [1,2,3] — running max

# functools
functools.reduce(lambda a,b: a+b, [1,2,3,4])  # 10 — fold left
functools.lru_cache(maxsize=None)              # memoization decorator
functools.cmp_to_key(compare_fn)               # convert comparator to key function

# lru_cache for memoization in recursive functions
@functools.lru_cache(maxsize=None)
def dp(i, j):
    if i == 0: return j
    if j == 0: return i
    if s1[i-1] == s2[j-1]: return dp(i-1, j-1)
    return 1 + min(dp(i-1,j), dp(i,j-1), dp(i-1,j-1))
```

**Common mistakes:**
- `itertools` functions return iterators — wrap in `list()` to materialize
- `@lru_cache` requires all function arguments to be hashable — cannot cache functions with list arguments; convert to tuple
- `functools.reduce` raises `TypeError` on empty list — provide initial value as third argument
- `itertools.combinations` vs `permutations` — combinations ignore order, permutations don't

Problems: LC 46, LC 78, LC 39, LC 322 (lru_cache for DP)
