## §1 — Character Classification (`is*` methods)

All return `True` / `False`. Used heavily in sliding window and two pointer problems for filtering valid characters.

```python
s = "Hello123"

s.isalnum()     # True if all chars are letters OR digits — most common in LeetCode
s.isalpha()     # True if all chars are letters only
s.isdigit()     # True if all chars are digits — strict: no '.', '-', superscripts
s.isnumeric()   # True for digits + numeric chars (e.g. '²', '½') — broader than isdigit
s.isdecimal()   # True for decimal digits only — strictest of the three
s.isspace()     # True if all chars are whitespace
s.isupper()     # True if all cased chars are uppercase
s.islower()     # True if all cased chars are lowercase
s.istitle()     # True if first letter of each word is uppercase

# isdigit vs isnumeric vs isdecimal — edge cases
'123'.isdigit()     # True
'²'.isdigit()       # True  — superscript counts as digit
'²'.isnumeric()     # True
'²'.isdecimal()     # False — superscript not a decimal digit
'½'.isnumeric()     # True  — fraction counts as numeric
'½'.isdigit()       # False
# Safe rule: use isdigit() for standard 0-9 digit check in LeetCode
```

**Common mistakes:**
- `isalnum()` returns `False` for spaces — `' '.isalnum()` is `False`
- `isupper()` returns `False` if string has no cased characters — `'123'.isupper()` is `False`
- Use `isdigit()` not `isnumeric()` for standard digit checks — `isnumeric()` accepts unexpected characters

## §2 — Case Conversion

```python
s = "hello world"

s.upper()         # 'HELLO WORLD' — all uppercase
s.lower()         # 'hello world' — all lowercase
s.capitalize()    # 'Hello world' — first char upper, rest lower
s.title()         # 'Hello World' — first char of each word upper
s.swapcase()      # 'HELLO WORLD' → 'hello world' — flip each char's case

# Case-insensitive comparison
s1.lower() == s2.lower()      # compare ignoring case
s1.casefold() == s2.casefold() # more aggressive — handles unicode ligatures
```

**Common mistakes:**
- `capitalize()` lowercases everything after the first character — `'hELLO'.capitalize()` = `'Hello'`
- Strings are immutable — all methods return a new string; the original is unchanged

Problems: LC 412, LC 709

## §3 — Search & Detection

```python
s = "hello world"

# Find position
s.find('o')           # 4 — first occurrence, -1 if not found (safe)
s.rfind('o')          # 7 — last occurrence, -1 if not found
s.find('o', 5)        # 7 — search from index 5 onward
s.index('o')          # 4 — like find() but raises ValueError if not found
s.rindex('o')         # 7 — last occurrence, raises ValueError if not found

# Count
s.count('l')          # 3 — non-overlapping occurrences
s.count('ll')         # 1 — counts non-overlapping substrings

# Boolean checks
s.startswith('hel')   # True
s.startswith(('hel', 'world'))  # True — accepts tuple of prefixes
s.endswith('ld')      # True
s.endswith(('ld', 'lo'))        # True — accepts tuple of suffixes
'world' in s          # True — O(n) substring check
```

**Common mistakes:**
- `find()` vs `index()` — use `find()` in LeetCode to avoid uncaught `ValueError`
- `count()` counts non-overlapping — `'aaa'.count('aa')` returns `1` not `2`
- `in` operator is O(n) — fine for LeetCode but not for performance-critical production code

Problems: LC 28, LC 459, LC 686

## §4 — Modification & Cleaning

```python
s = "  hello world  "

# Strip whitespace
s.strip()             # 'hello world' — both ends
s.lstrip()            # 'hello world  ' — left only
s.rstrip()            # '  hello world' — right only
s.strip('xyz')        # strip specific characters from both ends (not substring)

# Replace
s.replace('l', 'L')           # replace ALL occurrences
s.replace('l', 'L', 2)        # replace at most 2 occurrences
s.replace('l', '')            # delete all 'l' characters

# Bulk character replacement — translate
table = str.maketrans('aeiou', 'AEIOU')   # map each char to another
s.translate(table)                         # apply mapping

# Delete characters using translate
table = str.maketrans('', '', '!@#$%')    # third arg = chars to delete
s.translate(table)                         # removes all punctuation in one pass
```

**Common mistakes:**
- `strip('abc')` strips any combination of 'a', 'b', 'c' from the ends — not the substring `'abc'`
- `replace()` returns a new string — always assign the result
- `str.maketrans` third argument deletes characters — much faster than looping for bulk removal

Problems: LC 844, LC 1544, LC 1221

## §5 — Split & Join

```python
s = "hello   world  python"

# Split
s.split()             # ['hello', 'world', 'python'] — splits on ANY whitespace, removes empty strings
s.split(' ')          # ['hello', '', '', 'world', '', 'python'] — splits on SINGLE space, keeps empty strings
s.split(',')          # split on specific separator
s.split(',', 2)       # maxsplit=2 — at most 2 splits
s.rsplit(',', 1)      # split from right, at most 1 split
s.splitlines()        # split on line boundaries (\n, \r\n, \r)

# Partition — split into exactly 3 parts: before, sep, after
s.partition('@')      # ('user', '@', 'domain.com') — sep always included
s.rpartition('.')     # split at LAST occurrence

# Join
''.join(['a', 'b', 'c'])       # 'abc' — no separator
', '.join(['a', 'b', 'c'])     # 'a, b, c'
' '.join(reversed(s.split()))  # reverse words in a sentence
' '.join(s.split())            # normalize multiple spaces to single space
```

**Key distinction — `split()` vs `split(' ')`:**
- `split()` with no argument: splits on any whitespace, collapses consecutive whitespace, strips leading/trailing
- `split(' ')` with space: splits on single space only, preserves empty strings between consecutive spaces
- **Always use `split()` for word tokenization in LeetCode**

**Common mistakes:**
- `join()` called on the separator, not the list — `' '.join(lst)` not `lst.join(' ')`
- `join()` requires all elements to be strings — use `''.join(str(x) for x in lst)` for mixed types
- Using `+` to concatenate in a loop — O(n²); always use `join()` for building strings

## §6 — Formatting & Padding

```python
# Padding
'42'.zfill(5)             # '00042' — zero-pad a numeric string
'hi'.ljust(10)            # 'hi        ' — left-align, pad right with spaces
'hi'.rjust(10)            # '        hi' — right-align, pad left with spaces
'hi'.center(10)           # '    hi    ' — center with spaces
'hi'.ljust(10, '-')       # 'hi--------' — custom fill character

# f-string formatting
val = 3.14159
f'{val:.2f}'              # '3.14' — 2 decimal places
f'{val:10.2f}'            # '      3.14' — width 10, 2 decimal places
f'{42:05d}'               # '00042' — zero-padded integer
f'{42:>10}'               # '        42' — right-align in width 10
f'{42:<10}'               # '42        ' — left-align
f'{42:^10}'               # '    42    ' — center
f'{255:b}'                # '11111111' — binary representation
f'{255:x}'                # 'ff' — hex representation
f'{255:o}'                # '377' — octal representation
```

**Common mistakes:**
- `zfill()` pads with zeros specifically — use `ljust`/`rjust` for other fill characters
- f-string format spec order: `{value:fill align width .precision type}` — width before precision

Problems: LC 67, LC 190 (binary representation)

## §7 — Character ↔ Integer (`ord`, `chr`)

```python
# ord: character → ASCII/Unicode integer
ord('a')    # 97
ord('z')    # 122
ord('A')    # 65
ord('Z')    # 90
ord('0')    # 48
ord('9')    # 57

# chr: integer → character
chr(97)     # 'a'
chr(65)     # 'A'
chr(48)     # '0'

# Common patterns in LeetCode
ord(ch) - ord('a')            # 0-indexed position of lowercase letter (a=0, b=1, ...)
chr(ord('a') + i)             # convert index back to letter

# Frequency array for lowercase letters
freq = [0] * 26
for ch in s:
    freq[ord(ch) - ord('a')] += 1

# Caesar cipher shift
def shift(ch, k):
    return chr((ord(ch) - ord('a') + k) % 26 + ord('a'))

# Digit character to integer
int('7')          # 7 — preferred
ord('7') - ord('0')  # 7 — alternative without int()
```

**Common mistakes:**
- Forgetting `% 26` in circular shifts — without it, shifted chars go outside a-z range
- Using `ord(ch) - ord('a')` on uppercase letters — result is wrong; normalize case first
- `ord()` accepts only single character — `ord('ab')` raises `TypeError`

## §8 — Slicing & Indexing

```python
s = 'hello'

# Basic indexing
s[0]          # 'h' — first character
s[-1]         # 'o' — last character
s[-2]         # 'l' — second to last

# Slicing s[start:stop:step] — stop is exclusive
s[1:4]        # 'ell' — index 1 to 3
s[:3]         # 'hel' — from start to index 2
s[2:]         # 'llo' — from index 2 to end
s[:]          # 'hello' — full copy
s[::2]        # 'hlo' — every other character
s[::-1]       # 'olleh' — reverse string
s[4:1:-1]     # 'oll' — reverse slice with step -1

# Checking palindrome
s == s[::-1]              # True if palindrome

# Substring extraction
s[i:j]                    # substring from i to j-1 (same as SUBSTRING in SQL)

# String is immutable — cannot assign to index
# s[0] = 'H'  → TypeError
# Convert to list, modify, join back
lst = list(s)
lst[0] = 'H'
s = ''.join(lst)
```

**Common mistakes:**
- `s[i:j]` stop index is exclusive — `s[1:4]` gives characters at index 1, 2, 3
- Negative step requires `start > stop` — `s[1:4:-1]` returns empty string
- Strings are immutable — convert to list for in-place character modification

Problems: LC 344, LC 541, LC 557

## §9 — `string` Module Constants

```python
import string

string.ascii_lowercase    # 'abcdefghijklmnopqrstuvwxyz'
string.ascii_uppercase    # 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
string.ascii_letters      # lowercase + uppercase
string.digits             # '0123456789'
string.hexdigits          # '0123456789abcdefABCDEF'
string.octdigits          # '01234567'
string.punctuation        # '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
string.whitespace         # ' \t\n\r\x0b\x0c'
string.printable          # all printable characters

# Common usage patterns
set(string.ascii_lowercase)   # O(1) lookup for letter check
ch in string.punctuation      # check if char is punctuation

# Build frequency dict using string.ascii_lowercase
freq = {ch: 0 for ch in string.ascii_lowercase}
```

## §10 — Common LeetCode String Idioms

**Common mistakes:**
- Building strings with `+` in a loop — O(n²); always accumulate in a list and join at the end
- Forgetting strings are immutable — every modification creates a new string
- Using `sorted(s1) == sorted(s2)` when Counter is sufficient — both work but Counter is O(n) vs O(n log n)
- Rotation check: `s1 in s2 + s2` only works when `len(s1) == len(s2)` — always check length first
