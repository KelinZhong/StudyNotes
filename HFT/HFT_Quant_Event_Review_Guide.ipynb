## 1. Probability Puzzles and Mental Math

Priority: highest. The single most likely area to be tested in a casual setting. Brain teasers dropped mid-conversation are the norm.

### Bayes' Theorem

| Section | Content |
|---|---|
| **What is this for?** | Updating a belief when you receive new evidence. |
| **Brief Introduction** | Bayes' Theorem reverses conditional probabilities: given that you observed evidence B, what is the probability of underlying cause A? It shows up constantly as a brain teaser, often disguised as a medical test or coin-selection problem. |
| **Key Formulas** | $P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}$ <br><br> Total probability: $P(B) = P(B \mid A) P(A) + P(B \mid A^c) P(A^c)$ <br><br> Notation: $P(A)$ = prior, $P(A \mid B)$ = posterior, $P(B \mid A)$ = likelihood |
| **Intuition** | A rare disease with a 99% accurate test still produces mostly false positives if only 1% of people have it, because the base rate dominates. Always anchor on the prior before updating. |
| **If they ask you...** | "A coin is fair (50%) or double-headed (50%). You flip heads. Probability it is the double-headed coin?" <br> $(1.0 \times 0.5) / (1.0 \times 0.5 + 0.5 \times 0.5) = 2/3$. |

### Expected Value and Linearity of Expectation

| Section | Content |
|---|---|
| **What is this for?** | Computing the average outcome of a random process -- the workhorse of every probability puzzle. |
| **Brief Introduction** | $E[X]$ is the probability-weighted average of all outcomes. Linearity of expectation ($E[X+Y] = E[X] + E[Y]$, always, even when X and Y are dependent) is the single most powerful shortcut for solving interview puzzles quickly. |
| **Key Formulas** | Discrete: $E[X] = \sum_i x_i P(X = x_i)$ <br><br> Continuous: $E[X] = \int x f(x) dx$ <br><br> Linearity: $E[aX + bY] = aE[X] + bE[Y]$ (always holds) <br><br> Product: $E[XY] = E[X]E[Y]$ only if X, Y independent |
| **Intuition** | Break any complex problem into simple indicator variables. "Expected number of fixed points in a random permutation?" -- define $X_i = 1$ if element i is in its own position. $E[X] = \sum E[X_i] = n \cdot (1/n) = 1$. |
| **If they ask you...** | "You roll a die. I pay you the face value in dollars. What is the fair price?" <br> $E[X] = (1+2+3+4+5+6)/6 = 3.50$. |

### Conditional Probability

| Section | Content |
|---|---|
| **What is this for?** | Computing probability when you have partial information. |
| **Brief Introduction** | $P(A \mid B)$ is the probability of A given that B has already occurred. It is the foundation of Bayes' Theorem and the source of many interview traps -- the "boy or girl paradox" depends entirely on the conditioning statement. |
| **Key Formulas** | $P(A \mid B) = \frac{P(A \cap B)}{P(B)}$ <br><br> Law of Total Probability: $P(A) = \sum_i P(A \mid B_i) P(B_i)$ <br><br> Independence test: $P(A \mid B) = P(A)$ iff A, B independent |
| **Intuition** | Conditioning shrinks the sample space. You are no longer asking "what can happen?" but "what can happen given what I already know?" Many puzzles hinge on carefully defining what you are conditioning on. |
| **If they ask you...** | "I have two children. One is a boy. Probability both are boys?" <br> Sample space {BB, BG, GB} (not GG). P(BB) = 1/3, not 1/2. The trap: "one is a boy" is not the same as "the first child is a boy." |

### Geometric and Negative Binomial Distribution

| Section | Content |
|---|---|
| **What is this for?** | Modeling "how many trials until success?" -- common in order-fill and retry problems. |
| **Brief Introduction** | The geometric distribution counts the number of independent Bernoulli trials until the first success. The negative binomial generalizes to r successes. In HFT, these model things like "how many quotes until a fill?" |
| **Key Formulas** | Geometric: $P(X=k) = (1-p)^{k-1} p$ <br><br> $E[X] = 1/p$ , $\text{Var}(X) = (1-p)/p^2$ <br><br> Negative Binomial (r successes): $E[X] = r/p$ <br><br> Memoryless property: $P(X > s+t \mid X > s) = P(X > t)$ |
| **Intuition** | If a coin has a 10% chance of heads, you expect to flip it $1/0.1 = 10$ times for the first heads. The memoryless property means "the coin does not remember past failures" -- each trial is fresh. |
| **If they ask you...** | "Expected number of die rolls to see a 6?" <br> $p = 1/6$, so $E[X] = 6$ rolls. |

### Coupon Collector Problem

| Section | Content |
|---|---|
| **What is this for?** | "How many tries to collect all n distinct types?" -- models coverage and completeness problems. |
| **Brief Introduction** | You draw coupons uniformly at random from n types. On average, how many draws until you have one of each? It is a sum of geometric random variables with decreasing success probabilities. Very common puzzle pattern. |
| **Key Formulas** | $E[T] = n \sum_{k=1}^{n} \frac{1}{k} = n \cdot H_n \approx n \ln n + 0.577n$ <br><br> $H_n$ is the n-th harmonic number. <br><br> After collecting k types, $P(\text{new type}) = (n-k)/n$ <br><br> Expected draws for next new type $= n/(n-k)$ |
| **Intuition** | The first coupon is always new. When you have n-1 out of n, each draw has only a 1/n chance of completing the set -- the last coupon takes about n draws on its own. Most of the time is spent hunting the last few types. |
| **If they ask you...** | "Expected rolls of a fair die to see all 6 faces?" <br> $6(1 + 1/2 + 1/3 + 1/4 + 1/5 + 1/6) = 6 \times 2.45 = 14.7$ rolls. |

### Gambler's Ruin

| Section | Content |
|---|---|
| **What is this for?** | Probability of hitting a target vs. going broke in a random walk -- models profit targets and stop-losses. |
| **Brief Introduction** | A gambler starts with k dollars and bets 1 dollar per round (win prob p, lose prob q = 1-p). What is the probability of reaching n dollars before going to 0? This maps directly to barrier-hitting in random walks, relevant to position limits and drawdown analysis. |
| **Key Formulas** | Fair game ($p = q = 1/2$): $P(\text{reach } n) = k/n$ <br><br> Unfair game ($p \ne q$): $P(\text{reach } n) = \frac{1 - (q/p)^k}{1 - (q/p)^n}$ <br><br> Expected duration (fair): $E[T] = k(n-k)$ |
| **Intuition** | Even in a fair game, a player with fewer chips is more likely to go broke against a richer opponent. In HFT terms: a small edge repeated many times compounds, but capital constraints create absorbing barriers. |
| **If they ask you...** | "You have 20 dollars, target is 100 dollars, fair coin flip +/- 1 dollar. Probability of reaching 100?" <br> $20/100 = 20\%$. |

### Mental Math Shortcuts

| Section | Content |
|---|---|
| **What is this for?** | Speed and confidence under social pressure -- quick calculations without pen and paper. |
| **Brief Introduction** | Quant events often test mental math: fast multiplication, division, percentage estimation. The goal is not perfection but reasonable accuracy with speed. Being visibly comfortable with numbers signals quantitative fluency. |
| **Key Techniques** | Multiply by 99: $x \times 99 = x \times 100 - x$ <br><br> Squares near 50: $52^2 = 2500 + 100(2) + 4 = 2704$ <br><br> Divide by 7: $1/7 \approx 0.143$, repeating cycle 142857 <br><br> Percentage flips: 8% of 25 = 25% of 8 = 2 <br><br> Rule of 72: doubling time $\approx 72 / r\%$ |
| **Intuition** | Break hard computations into easy ones. $17 \times 23 = (20-3)(20+3) = 400 - 9 = 391$. Anchor to nearby round numbers and adjust. |
| **If they ask you...** | "What is $37 \times 43$?" <br> $(40-3)(40+3) = 1600 - 9 = 1591$. Show the method, not just the answer. |

### Fermi Estimation

| Section | Content |
|---|---|
| **What is this for?** | Estimating unknowable quantities to the right order of magnitude -- tests structured thinking. |
| **Brief Introduction** | Fermi questions ask you to estimate something you cannot look up: "How many piano tuners are in Chicago?" The answer does not matter -- what matters is your decomposition into estimable sub-problems and your comfort with rough numbers. |
| **Key Techniques** | Decompose: break into factors you can estimate independently. <br><br> Anchor: start from a known quantity (population, time in a day, etc.). <br><br> Bound: give a range, then narrow -- "definitely more than X, less than Y, so probably around Z." <br><br> Cross-check: verify with an independent approach. |
| **Intuition** | The power comes from multiplication: even if each factor is off by 2x, three factors multiplied give only about 8x error -- usually within the right order of magnitude. |
| **If they ask you...** | "How many trades does the NYSE process per day?" <br> About 3,000 listed stocks times about 10,000 trades per stock per day (active stocks more, inactive less) gives about 30 million. Actual is roughly 30-50M. |

### Common Brain Teasers

| # | Puzzle | Quick Answer | Brief Process |
|---|---|---|---|
| 1 | Expected value of max of two dice? | $161/36 \approx 4.47$ | $E[\max] = \sum_{k=1}^{6} P(\max \ge k)$. Compute $P(\max \ge k) = 1 - P(\text{both} < k) = 1 - ((k-1)/6)^2$. Sum over k = 1 to 6. |
| 2 | Flip a fair coin until HH. Expected flips? | 6 flips | Set up Markov states: S0 (start), S1 (just got H), S2 (done). $E_0 = 1 + 0.5 E_0 + 0.5 E_1$; $E_1 = 1 + 0.5 E_0$. Solve: $E_0 = 6$. |
| 3 | Two envelopes: one has 2x the other. Switch? | No advantage | The "paradox" arises from incorrect conditioning -- you cannot simultaneously assume your envelope is the smaller with prob 1/2 across all values. No information is gained by opening, so switching has zero expected benefit. |
| 4 | 100 people board a plane, first takes a random seat. Probability last person gets their own seat? | 1/2 | At each step the displaced person either sits in seat 1 (game ends, last person gets own seat) or seat 100 (game ends, last person does not), or displaces someone else. By symmetry, seats 1 and 100 are equally likely to be the last filled. |
| 5 | Sum of integers 1 to 100? | 5050 | Gauss: pair first and last terms. $(1+100) + (2+99) + \ldots = 50 \times 101 = 5050$. General: $n(n+1)/2$. |
| 6 | Ways to climb n stairs taking 1 or 2 steps? | Fibonacci | To reach step n, you came from step n-1 (1 step) or n-2 (2 steps). So $F(n) = F(n-1) + F(n-2)$, with $F(1) = 1$, $F(2) = 2$. |
| 7 | Expected people until two share a birthday (n=365)? | About 24 | Birthday problem: $P(\text{match in k people}) = 1 - \prod_{i=0}^{k-1}(1 - i/365)$. Crosses 50% at k = 23. Expected value (first match) $\approx \sqrt{\pi n / 2} \approx 24$. |
| 8 | Monty Hall: switch or stay? | Switch, 2/3 | Initial pick has 1/3 chance. Monty reveals a goat (he always can). The remaining door absorbs the other 2/3. Switching wins whenever your initial pick was wrong, which happens 2/3 of the time. |
| 9 | Break a stick at two random points. Probability the three pieces form a triangle? | 1/4 | Let the two break points be $U, V$ uniform on [0,1]. Need all three pieces < 1/2 (triangle inequality). The feasible region is 1/4 of the unit square. |
| 10 | Draw cards from a standard deck. Expected draws until the first Ace? | 10.6 | By symmetry, 4 Aces divide 48 non-Aces into 5 equally likely gaps. Expected size of the first gap = 48/5 = 9.6 non-Aces, plus 1 for the Ace itself = 10.6. Equivalently, $E = 53/5$. |

## 2. Probability and Statistics Foundations

Priority: high. Know these well enough to explain them in 30 seconds at a cocktail table. Focus on why over how to compute.

### Law of Large Numbers (LLN)

| Section | Content |
|---|---|
| **What is this for?** | Why HFT works: a small edge converges to expected profit over millions of trades. |
| **Brief Introduction** | LLN says the sample average $\bar{X}_n$ converges to the true mean $\mu$ as $n \to \infty$. In HFT, your edge per trade might be tiny, but repeating it thousands of times a day makes the aggregate P&L predictable. |
| **Key Formulas** | Weak LLN: $\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i \xrightarrow{P} \mu$ as $n \to \infty$ <br><br> Requires: $X_i$ i.i.d. with finite mean. |
| **Intuition** | Casinos do not gamble -- they play millions of hands. The house edge is tiny per hand but certain in aggregate. An HFT market maker is the casino. |
| **If they ask you...** | "Why does a strategy with 51% win rate make money?" <br> LLN. Over enough trades, the sample win rate converges to 51%, guaranteeing profit. The key assumption: trades are sufficiently independent. |

### Central Limit Theorem (CLT)

| Section | Content |
|---|---|
| **What is this for?** | Why the normal distribution appears everywhere -- and why it sometimes does not. |
| **Brief Introduction** | CLT says the sum or average of many independent random variables converges to a normal distribution, regardless of the underlying distribution. This justifies using normal-based tools (z-scores, confidence intervals) but fails when independence breaks down or distributions have infinite variance. |
| **Key Formulas** | $\frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}} \xrightarrow{d} N(0, 1)$ <br><br> Convergence rate: $O(1/\sqrt{n})$ -- need 4x data for 2x precision. |
| **Intuition** | Roll one die: uniform. Roll 100 dice and average: nearly bell-shaped. The magic is that the shape of the original distribution washes out. The warning: CLT breaks in the tails -- exactly where risk lives. |
| **If they ask you...** | "Are stock returns normally distributed?" <br> Approximately in the center, but tails are fatter than normal (leptokurtic). CLT holds for aggregated returns over longer horizons, but tick-by-tick returns deviate significantly. |

### Variance, Covariance, and Correlation

| Section | Content |
|---|---|
| **What is this for?** | Measuring risk (variance), co-movement (covariance), and standardized co-movement (correlation). |
| **Brief Introduction** | Variance measures spread around the mean. Covariance captures the direction of linear co-movement between two variables. Correlation normalizes covariance to [-1, 1], making it comparable across assets. These are the language of risk and diversification. |
| **Key Formulas** | $\text{Var}(X) = E[X^2] - (E[X])^2$ <br><br> $\text{Cov}(X,Y) = E[XY] - E[X]E[Y]$ <br><br> $\rho_{X,Y} = \text{Cov}(X,Y) / (\sigma_X \sigma_Y)$, range [-1, 1] <br><br> $\text{Var}(X+Y) = \text{Var}(X) + \text{Var}(Y) + 2\text{Cov}(X,Y)$ |
| **Intuition** | Correlation = 0 means no linear relationship -- X and Y could still be dependent (e.g. $Y = X^2$). In a crisis, correlations spike to 1 -- diversification fails exactly when you need it. |
| **If they ask you...** | "Two assets both have 20% vol and 0.5 correlation. Portfolio vol at 50/50 weight?" <br> $\sigma_p = \sqrt{0.25(0.04) + 0.25(0.04) + 2(0.25)(0.5)(0.2)(0.2)} = \sqrt{0.03} \approx 17.3\%$. |

### Fat Tails and Skewness

| Section | Content |
|---|---|
| **What is this for?** | Understanding why real financial data is more extreme than the normal distribution predicts. |
| **Brief Introduction** | Financial returns exhibit leptokurtosis (fat tails) -- extreme events happen far more often than a Gaussian model predicts. Returns are also often negatively skewed (big losses more common than big gains). Ignoring this misprices risk catastrophically. |
| **Key Formulas** | Skewness: $S = E[(X-\mu)^3] / \sigma^3$ -- measures asymmetry <br> $S < 0$: left tail heavier; $S > 0$: right tail heavier <br><br> Kurtosis: $K = E[(X-\mu)^4] / \sigma^4$ -- measures tail weight (Normal $K = 3$) <br><br> Excess Kurtosis: $K - 3$ (positive = fatter tails than normal) |
| **Intuition** | The 2010 Flash Crash, COVID March 2020 -- these are "25-sigma events" under a normal model, meaning they should occur once per billions of years. They happen every few years because the true distribution has fat tails. |
| **If they ask you...** | "Are you comfortable using normal distributions for risk?" <br> For central tendency and portfolio construction, yes. For tail risk (VaR, drawdown), absolutely not -- use empirical distributions, Student-t, or extreme value theory. |

### Common Distributions -- When Each Appears in HFT

| Distribution | Key Parameters | E[X] | Var(X) | HFT Use Case |
|---|---|---|---|---|
| Normal $N(\mu, \sigma^2)$ | mean $\mu$, variance $\sigma^2$ | $\mu$ | $\sigma^2$ | Returns modeling (central), portfolio risk |
| Poisson $\text{Pois}(\lambda)$ | rate $\lambda$ | $\lambda$ | $\lambda$ | Order arrival counts per time window |
| Exponential $\text{Exp}(\lambda)$ | rate $\lambda$ | $1/\lambda$ | $1/\lambda^2$ | Time between order arrivals (memoryless) |
| Uniform $U(a,b)$ | lower $a$, upper $b$ | $(a+b)/2$ | $(b-a)^2/12$ | Random order placement within a range |
| Bernoulli $\text{Ber}(p)$ | success prob $p$ | $p$ | $p(1-p)$ | Win/loss per trade, fill/no-fill |
| Binomial $\text{Bin}(n,p)$ | trials $n$, prob $p$ | $np$ | $np(1-p)$ | Number of fills out of n attempts |

## 3. Market Microstructure

Priority: highest. This is the domain knowledge that shows you understand the business, not just the math.

### Bid-Ask Spread and Mid Price

| Section | Content |
|---|---|
| **What is this for?** | The fundamental unit of HFT economics -- the cost of immediacy. |
| **Brief Introduction** | The bid is the highest price a buyer will pay; the ask is the lowest price a seller will accept. The spread between them is the market maker's revenue per round-trip and the trader's cost of execution. In HFT, fractions of a cent matter. |
| **Key Formulas** | Spread $= P_{\text{ask}} - P_{\text{bid}}$ <br><br> Mid Price $= (P_{\text{ask}} + P_{\text{bid}}) / 2$ <br><br> Effective Spread $= 2 \times \lvert P_{\text{trade}} - P_{\text{mid}} \rvert$ <br><br> Relative Spread $= (P_{\text{ask}} - P_{\text{bid}}) / P_{\text{mid}}$ |
| **Intuition** | Tight spread = liquid, competitive market. Wide spread = risky or illiquid. The spread is not free money -- it is compensation for adverse selection, inventory risk, and operational cost. |
| **If they ask you...** | "How does a market maker make money?" <br> Post bid at 100 and ask at 100.02. Buy from a seller, sell to a buyer, capture 0.02. Profit if the fills are balanced and the price does not move against you before you close. |

### Order Book Mechanics

| Section | Content |
|---|---|
| **What is this for?** | Understanding the fundamental data structure of electronic markets. |
| **Brief Introduction** | The limit order book (LOB) is a queue of resting buy and sell orders at different price levels, sorted by price-time priority. Market orders consume liquidity from the book; limit orders provide it. The shape and dynamics of the order book drive price formation. |
| **Key Concepts** | Limit order: rests on the book at a specified price, provides liquidity, earns rebate. <br><br> Market order: executes immediately at the best available price, takes liquidity, pays fee. <br><br> Price-time priority (FIFO): first limit order at a price gets filled first. <br><br> Book depth: total quantity resting at each price level. <br><br> Book imbalance: $(Q_{\text{bid}} - Q_{\text{ask}}) / (Q_{\text{bid}} + Q_{\text{ask}})$ -- predictive of short-term price direction. |
| **Intuition** | The order book is a supply-demand curve in real time. Heavy buying interest (deep bids, thin asks) signals upward pressure. Book imbalance is one of the simplest and most robust short-term signals in HFT. |
| **If they ask you...** | "What is the difference between a limit and market order?" <br> A limit order says "I want to buy, but only at this price or better" (patient, earns spread). A market order says "I want to buy right now at whatever price" (impatient, crosses spread). |

### Adverse Selection

| Section | Content |
|---|---|
| **What is this for?** | Understanding why market making is hard -- and why the spread must exist. |
| **Brief Introduction** | Informed traders (who have better information about future prices) trade against market makers only when it is profitable to do so -- buying when the price is about to rise, selling when it is about to fall. The market maker systematically loses on these trades. The spread is compensation for bearing this "adverse selection" risk. |
| **Intuition** | Imagine selling used cars. Most buyers are normal, but some are mechanics who only buy when the car is underpriced. You widen your markup to survive the mechanics. That markup is the spread. The more informed flow, the wider the spread needs to be. |
| **If they ask you...** | "Why can't a market maker just set a zero spread?" <br> They would be picked off by informed traders on every move. The spread must cover the expected loss from adverse selection. Speed helps: faster market makers can cancel stale quotes before informed traders hit them. |

### Market Impact

| Section | Content |
|---|---|
| **What is this for?** | Why large orders move prices -- and how to minimize the cost of trading. |
| **Brief Introduction** | Market impact is the price movement caused by the act of trading. A large buy order consumes liquidity at each price level, walking up the order book. This is the dominant cost for institutional traders and the dominant revenue source for market makers facing informed flow. |
| **Key Formulas** | Square-root law (empirical): <br> Impact $\approx \sigma \sqrt{Q/V}$ <br><br> where $\sigma$ = volatility, $Q$ = order size, $V$ = daily volume. <br><br> Key insight: impact scales as the square root of order size -- doubling your order does not double your cost. |
| **Intuition** | Dropping a stone in a pond creates ripples proportional to the stone's size. But the ripples spread out, so a 2x bigger stone creates roughly $\sqrt{2}$x bigger waves, not 2x. Similarly, trading 4x more shares costs about 2x more per share. |
| **If they ask you...** | "Why not just send one big order?" <br> You would pay enormous market impact. Better to slice the order into smaller child orders spread over time (TWAP/VWAP), letting the book refill between executions. |

### Latency and Co-location

| Section | Content |
|---|---|
| **What is this for?** | Understanding the technology edge that defines HFT. |
| **Brief Introduction** | In HFT, speed is alpha. Latency is the time delay between observing a market event and acting on it. Co-location means placing your servers physically next to the exchange's matching engine to minimize this delay. Firms compete on microseconds. |
| **Key Concepts** | Co-location: servers in the same data center as the exchange (roughly 5-50 microseconds latency). <br><br> FPGA: field-programmable gate arrays for hardware-level speed (bypass the OS entirely). <br><br> Kernel bypass: software techniques to skip the OS network stack. <br><br> Tick-to-trade: total time from receiving market data to sending an order -- top firms achieve sub-microsecond. |
| **Intuition** | Light travels about 300m per microsecond. A firm in New Jersey has a roughly 5ms round-trip to Chicago. If your competitor is co-located and you are not, they see every opportunity before you do. Speed advantages are self-reinforcing: faster firms capture more spread, earn more revenue, invest more in speed. |
| **If they ask you...** | "Why does speed matter if your signal is better?" <br> Because signals decay. A perfect signal that arrives 1ms late is worthless if someone already traded on it. Latency determines whether you capture or miss the opportunity, especially for cross-venue arbitrage. |

### Maker-Taker Fee Models

| Section | Content |
|---|---|
| **What is this for?** | Understanding how exchange fee structures create incentives that shape HFT strategy. |
| **Brief Introduction** | Most exchanges use a maker-taker model: limit orders that provide liquidity (makers) receive a rebate; market orders that take liquidity pay a fee. This incentivizes passive market making and affects where and how HFT firms route orders. |
| **Key Concepts** | Maker rebate: roughly 0.001-0.003 dollars per share (you earn this for resting limit orders that get filled). <br><br> Taker fee: roughly 0.003-0.005 dollars per share (you pay this for crossing the spread). <br><br> Inverted venues: flip the model -- pay makers, rebate takers -- to attract aggressive flow. <br><br> Net spread = quoted spread minus maker rebate (effective revenue for market makers). |
| **Intuition** | Exchanges compete for order flow by adjusting rebates. A market maker earning 0.01 spread plus 0.002 rebate makes 0.012 per share. If the rebate disappears, the strategy might not be profitable. Fee structures are a core input to strategy P&L. |
| **If they ask you...** | "Do exchange fees matter at HFT scale?" <br> Absolutely. At millions of trades per day, a 0.001 dollar per share difference in fees is millions of dollars annually. Some strategies are only profitable because of rebates. |

## 4. Core Quant Concepts

Priority: high. Things a researcher might casually ask: "So, what do you know about...".

### Mean Reversion vs. Momentum

| Section | Content |
|---|---|
| **What is this for?** | The two fundamental signal families in quantitative trading. |
| **Brief Introduction** | Mean reversion bets that prices overshoot and snap back ("buy the dip"). Momentum bets that trends continue ("buy the breakout"). Most HFT strategies are mean-reverting at short timescales (microstructure noise) and potentially momentum-following at longer timescales. |
| **Key Concepts** | Mean Reversion: negative autocorrelation in returns -- if the price went up, it is more likely to come back down. Common in market-making, pairs trading. <br><br> Momentum: positive autocorrelation -- if the price went up, it is more likely to keep going up. Common in trend-following, CTA strategies. <br><br> Timescale dependence: the same asset can be mean-reverting at 1-second and trending at 1-month. |
| **Intuition** | Imagine a drunk person walking a dog on a leash. The drunk (fundamental value) wanders randomly -- that is momentum/random walk. The dog (price) oscillates around the drunk -- that is mean reversion around a moving anchor. The leash length is the spread. |
| **If they ask you...** | "Is the market mean-reverting or trending?" <br> Both, depending on timescale. HFT exploits mean reversion (bid-ask bounce, temporary dislocations). Macro strategies exploit momentum (trends in rates, currencies). The smart answer acknowledges both. |

### Sharpe Ratio

| Section | Content |
|---|---|
| **What is this for?** | The standard metric for strategy quality -- risk-adjusted return. |
| **Brief Introduction** | The Sharpe ratio measures return per unit of risk (standard deviation). It is the first number anyone asks about a strategy. It allows comparison across strategies with different leverage and volatility. |
| **Key Formulas** | $\text{Sharpe} = (E[R_p] - R_f) / \sigma_p$ <br><br> Annualized: $\text{Sharpe}_{\text{ann}} = \text{Sharpe}_{\text{daily}} \times \sqrt{252}$ <br><br> Note: $\text{Sharpe}_{\text{daily}} \times \sqrt{252}$ overestimates if returns are autocorrelated. |
| **Intuition** | Sharpe = 1 means 1% excess return per 1% volatility. Benchmarks: below 0.5 weak, 0.5-1.0 decent, 1.0-2.0 good, 2.0-3.0 excellent, above 3.0 check for errors or overfitting. HFT strategies often have Sharpe above 5 due to high frequency and small per-trade risk. |
| **If they ask you...** | "What is a good Sharpe ratio?" <br> Depends on frequency. A long-only equity fund at Sharpe 1.0 is excellent. An HFT strategy at Sharpe 1.0 is terrible -- you are taking thousands of bets daily, so the CLT should give you very high Sharpe. If it is only 1.0, the edge is probably noise. |

### Overfitting and Data Snooping

| Section | Content |
|---|---|
| **What is this for?** | The single biggest risk in quantitative research -- finding patterns that do not exist. |
| **Brief Introduction** | Overfitting means your model has learned the noise in historical data rather than the true signal. Data snooping happens when you test many hypotheses on the same dataset and cherry-pick the ones that "worked." Both are lethal: the strategy looks great in backtest and dies in live trading. |
| **Key Concepts** | In-sample vs. out-of-sample: always validate on unseen data. <br><br> Multiple testing correction: if you test 100 signals, 5 will be "significant" at p < 0.05 by chance -- apply Bonferroni or BH correction. <br><br> Degrees of freedom: more parameters = more room to overfit. <br><br> Walk-forward validation: train on [0, T], test on [T, T+delta], roll forward. |
| **Intuition** | Fitting a polynomial through 10 data points perfectly requires a degree-9 polynomial -- but it tells you nothing about the 11th point. A good model is one that predicts new data well, not one that memorizes old data. |
| **If they ask you...** | "How do you know a signal is real?" <br> Out-of-sample testing, economic intuition (why should this work?), robustness to parameter changes, and statistical corrections for multiple testing. If a signal only works with one specific parameter set, it is probably overfit. |

### Correlation vs. Causation

| Section | Content |
|---|---|
| **What is this for?** | Avoiding the most common analytical mistake -- assuming a statistical relationship implies a causal one. |
| **Brief Introduction** | Two variables can be correlated because: (a) X causes Y, (b) Y causes X, (c) a third variable Z causes both, or (d) pure coincidence. In quant research, spurious correlations are everywhere -- your job is to distinguish signal from noise. |
| **Intuition** | Ice cream sales and drowning deaths are correlated. Neither causes the other -- summer heat causes both. In trading: two stocks might be correlated because they are in the same sector (Z), not because one's movement predicts the other. |
| **If they ask you...** | "If two assets are correlated, does a move in one predict the other?" <br> Not necessarily. You need to understand why they are correlated. If it is a common factor (sector, macro), the correlation is contemporaneous, not predictive. Only if one leads the other (e.g. due to different trading hours or information arrival) is there a tradeable signal. |

### Stationarity

| Section | Content |
|---|---|
| **What is this for?** | The fundamental requirement for time-series modeling -- and why you cannot just forecast a random walk. |
| **Brief Introduction** | A time series is stationary if its statistical properties (mean, variance, autocorrelation) do not change over time. Most raw price series are non-stationary (they trend). Returns (log-differences) are typically stationary. You cannot apply regression or correlation to non-stationary series without getting meaningless results (spurious regression). |
| **Key Concepts** | Stationary: returns, spreads between cointegrated pairs. <br><br> Non-stationary: raw prices, cumulative P&L. <br><br> Fix: take differences, use returns instead of prices, or model the cointegrating relationship. <br><br> Test: Augmented Dickey-Fuller (ADF) -- null hypothesis is "unit root" (non-stationary). |
| **Intuition** | Regressing Amazon's stock price on the number of iPhones sold over the past decade gives R-squared above 0.9. Meaningful? No -- both trended upward. Two random walks will show spurious correlation. Always work with stationary series. |
| **If they ask you...** | "Can you predict stock prices?" <br> No -- prices are approximately a random walk (non-stationary). You can predict returns (changes in price), which are approximately stationary. The distinction is fundamental. |

### Pairs Trading and Cointegration

| Section | Content |
|---|---|
| **What is this for?** | The classic statistical arbitrage strategy -- and the concept behind many HFT spread-trading signals. |
| **Brief Introduction** | Two assets are cointegrated if their prices wander individually (non-stationary) but the spread between them is stationary (mean-reverting). Pairs trading exploits this: when the spread widens, go long the cheap one, short the expensive one, and wait for convergence. |
| **Key Concepts** | Cointegration is not the same as correlation. Two assets can be correlated but not cointegrated (they trend in the same direction but the gap grows). Cointegration means the gap is bounded. <br><br> Spread: $S_t = P_t^A - \beta P_t^B$ where $\beta$ is the hedge ratio. <br><br> Trade signal: when $S_t$ deviates from mean by more than $k\sigma$, enter the trade; exit when it reverts. |
| **Intuition** | Think of two friends walking drunk on a street. Correlated means they stumble in the same direction. Cointegrated means they are tied together by a bungee cord -- they can diverge, but the cord always pulls them back. |
| **If they ask you...** | "Give me an example of a pairs trade." <br> Coca-Cola and Pepsi. Both driven by similar fundamentals, so their price ratio tends to be stable. If Coke drops 3% and Pepsi is flat with no news, you would go long Coke / short Pepsi and expect convergence. |

## 5. Derivatives Intuition

Priority: moderate. Only if the firm trades options or futures. Know the intuition, skip the derivations.

### Put-Call Parity

| Section | Content |
|---|---|
| **What is this for?** | The fundamental no-arbitrage relationship linking calls, puts, stock, and bonds. |
| **Brief Introduction** | Put-call parity states that a call and a put with the same strike and expiry, combined with a bond, replicate the underlying stock. It is not a model -- it is an identity. Any violation is a pure arbitrage opportunity. |
| **Key Formulas** | $C - P = S - K e^{-rT}$ <br><br> Equivalently: $C + Ke^{-rT} = P + S$ <br><br> Notation: C = call price, P = put price, S = stock price, K = strike, r = risk-free rate, T = time to expiry. |
| **Intuition** | Owning a call plus cash (to buy the stock at expiry) is identical to owning a put plus the stock itself. Both portfolios have the exact same payoff in every state of the world. If they are priced differently, free money exists. |
| **If they ask you...** | "If a call costs 5, stock is 100, strike is 100, rate is 0, time is 1 year -- what is the put worth?" <br> $P = C - S + Ke^{-rT} = 5 - 100 + 100 = 5$. Call = Put when $S = Ke^{-rT}$ (at-the-money forward). |

### The Greeks

| Greek | Symbol | Formula | What It Measures | Intuition |
|---|---|---|---|---|
| Delta | $\Delta$ | $\partial V / \partial S$ | Sensitivity of option value to underlying price | "How many shares is this option equivalent to?" ATM call is roughly 0.5. |
| Gamma | $\Gamma$ | $\partial^2 V / \partial S^2$ | Rate of change of Delta as underlying moves | "How fast does my hedge need updating?" Highest for ATM near expiry. |
| Vega | $\mathcal{V}$ | $\partial V / \partial \sigma$ | Sensitivity of option value to implied volatility | "How much do I make or lose if vol changes 1%?" Long options = long vega. |
| Theta | $\Theta$ | $\partial V / \partial t$ | Rate of time decay | "How much value do I lose per day?" Options are wasting assets -- theta is the cost of waiting. |
| Rho | $\rho$ | $\partial V / \partial r$ | Sensitivity of option value to interest rates | Usually small. Matters for long-dated options. |

**Notation:**
- $V$ = option value (price of the call or put).
- $S$ = current price of the underlying asset (e.g. stock price).
- $\sigma$ = implied volatility of the underlying, expressed as an annualized standard deviation of returns.
- $t$ = time; $\partial V / \partial t$ measures how V changes as time passes (with expiry approaching).
- $r$ = risk-free interest rate (annualized, continuously compounded).
- $\partial$ denotes a partial derivative: the rate of change of one variable while holding all others constant.
- ATM = at-the-money, meaning the option's strike price equals the current underlying price.

**Key relationship (Black-Scholes PDE):**

$\Theta + \frac{1}{2}\sigma^2 S^2 \Gamma + rS\Delta = rV$

This links Theta and Gamma directly: you cannot have convexity (positive Gamma) without paying for it (negative Theta). A long option position earns from large moves (Gamma) but bleeds value each day (Theta). The PDE says these forces must balance in equilibrium.

### Implied vs. Realized Volatility

| Section | Content |
|---|---|
| **What is this for?** | The core of volatility trading -- is the market over- or under-pricing risk? |
| **Brief Introduction** | Implied volatility (IV) is the market's forecast of future vol, extracted from option prices via Black-Scholes. Realized volatility (RV) is what actually happened. The gap between them is the volatility risk premium -- and the source of many trading strategies. |
| **Key Formulas** | Realized Vol $= \sqrt{(252/n) \sum_{i=1}^{n} r_i^2}$ <br> (annualized, assuming zero mean) <br><br> IV: the $\sigma$ that makes Black-Scholes price equal to market price. <br><br> Vol Risk Premium = IV - RV <br> (typically positive -- sellers of options earn this premium) |
| **Intuition** | IV is like the insurance premium the market charges for uncertainty. It is usually higher than what actually materializes (RV), because people overpay for protection. Selling options (short vol) captures this premium -- but blows up in tail events. |
| **If they ask you...** | "IV is 25% but you think RV will be 20%. What do you do?" <br> Sell options (short vol). You are collecting a premium based on 25% vol but the actual movement will only be 20%. Delta-hedge to isolate the pure vol bet. <br> P&L is roughly $\frac{1}{2} \Gamma S^2 (\sigma_{\text{implied}}^2 - \sigma_{\text{realized}}^2)$ per day. |

### Futures Basis and Cost of Carry

| Section | Content |
|---|---|
| **What is this for?** | Understanding why futures price does not equal spot price -- and why the gap is tradeable. |
| **Brief Introduction** | The futures price reflects the spot price plus the cost of holding ("carrying") the underlying until expiry: financing, storage, dividends foregone. The basis (futures minus spot) is normally positive (contango) but can go negative (backwardation). Deviations from fair value are arbitrage opportunities. |
| **Key Formulas** | $F_0 = S_0 \cdot e^{(r - d)T}$ <br><br> where r = risk-free rate, d = dividend yield, T = time to expiry. <br><br> Basis $= F_0 - S_0$ <br><br> Contango: $F > S$ (normal when $r > d$). <br> Backwardation: $F < S$ (common in commodities with convenience yield). |
| **Intuition** | Buying a stock costs money (financing at rate r) but earns dividends (d). The futures buyer does not hold the stock, so they must pay the net carry cost in the futures price. If the futures price deviates from this fair value, an arbitrageur will lock in the difference. |
| **If they ask you...** | "S&P 500 spot is 5000, rate is 5%, dividend yield is 1.5%, 3-month future -- what is the fair price?" <br> $F = 5000 \times e^{(0.05 - 0.015) \times 0.25} = 5000 \times e^{0.00875} \approx 5044$. |

## 6. Industry Knowledge and Smart Questions

Priority: highest. What separates someone who is prepared from someone who is just smart. This is not about formulas -- it is about showing genuine understanding of the business.

### What Is HFT, Really?

| Section | Content |
|---|---|
| **What is this for?** | Being able to explain HFT clearly -- because they will judge whether you understand their business. |
| **Brief Introduction** | HFT is a set of strategies that use speed (low-latency infrastructure) and statistical edge (quantitative models) to trade at high frequency. It is not a single strategy -- it is a technology-enabled approach to multiple strategy types. |
| **Key Strategy Types** | Market Making: continuously posting bid/ask quotes, earning the spread, managing inventory risk. Revenue = spread times volume minus adverse selection losses. <br><br> Statistical Arbitrage: exploiting temporary mispricings between related instruments (pairs, ETFs vs. baskets, futures vs. spot). <br><br> Latency Arbitrage: exploiting speed advantages to react to information before others (e.g. price changes on one exchange before another updates). <br><br> Event-Driven: ultra-fast reaction to news, earnings releases, economic data. |
| **If they ask you...** | "What does our firm actually do?" <br> Research beforehand. But generally: "You provide liquidity to the market through electronic market making and capture mispricings through statistical arbitrage, using proprietary low-latency infrastructure to minimize adverse selection and maximize fill rates." |

### How HFT Firms Make Money

| Section | Content |
|---|---|
| **What is this for?** | Understanding the P&L drivers -- essential for talking to practitioners. |
| **Revenue Sources** | Spread capture: earning the bid-ask spread on balanced flow. <br><br> Rebates: exchange maker-taker rebates for providing liquidity. <br><br> Alpha signals: predictive models that forecast short-term price movements (book imbalance, trade flow toxicity, cross-asset signals). <br><br> Arbitrage: capturing mispricings (ETF vs. basket, futures vs. spot, cross-exchange). |
| **Cost Drivers** | Adverse selection: losses from trading against informed flow. <br><br> Technology: co-location, FPGA, networking, data feeds -- millions per year. <br><br> Market data fees: exchange data is expensive. <br><br> Regulatory and compliance: market surveillance, reporting. |
| **If they ask you...** | "Is HFT just front-running?" <br> No. Front-running is illegal (trading ahead of a known client order). HFT firms react to public market data faster than others -- that is competition, not front-running. Market making actually improves liquidity and tightens spreads. |

### Regulatory Landscape

| Section | Content |
|---|---|
| **What is this for?** | Showing awareness that HFT operates under intense scrutiny. |
| **Key Concepts** | Spoofing/Layering: placing orders you intend to cancel to manipulate the order book -- this is illegal (Dodd-Frank). <br><br> Reg NMS (US): National Market System -- requires orders to be routed to the best available price across all exchanges, which created arbitrage opportunities across venues. <br><br> MiFID II (EU): requires algorithmic trading firms to register, test algos, and maintain kill switches. <br><br> Market-making obligations: some exchanges require registered market makers to quote continuously within spread limits. |
| **If they ask you...** | "What is spoofing?" <br> Placing large orders you never intend to execute, to create a false impression of supply or demand and trick other participants into trading at a disadvantageous price. It is illegal and firms have been fined hundreds of millions for it. |

### Technology Edge

| Section | Content |
|---|---|
| **What is this for?** | Vocabulary to discuss the engineering side without pretending to be a systems engineer. |
| **Key Concepts** | FPGA: custom hardware that processes market data in nanoseconds -- bypasses CPU entirely, used for simple speed-critical decisions. <br><br> Co-location: placing servers in the same building as the exchange's matching engine. <br><br> Kernel bypass: software that skips the operating system's network stack for faster I/O. <br><br> Tick-to-trade latency: time from receiving a market data tick to sending an order -- top firms achieve sub-microsecond. <br><br> C++ / Rust: languages of choice for latency-critical code (not Python). |
| **If they ask you...** | "Why C++ instead of Python?" <br> Python is roughly 100x slower for latency-critical paths. In HFT, a 10 microsecond delay means someone else captures the opportunity. Python is used for research, backtesting, and signal development; C++ for production execution. |
