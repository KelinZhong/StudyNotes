## Key Formulas -- At a Glance

| Concept | Formula | Quick Note |
|---|---|---|
| Bayes' Theorem | $P(A \mid B) = \frac{P(B \mid A) P(A)}{P(B)}$ | Prior times Likelihood over Evidence |
| Expected Value | $E[X] = \sum x_i P(x_i)$ | Linearity always holds: <br> $E[X+Y] = E[X]+E[Y]$ |
| Variance | $\text{Var}(X) = E[X^2] - (E[X])^2$ | $\text{Var}(aX) = a^2 \text{Var}(X)$ |
| Covariance | $\text{Cov}(X,Y) = E[XY] - E[X]E[Y]$ | Equals 0 if independent <br> (converse is false) |
| Correlation | $\rho = \text{Cov}(X,Y) / (\sigma_X \sigma_Y)$ | Range [-1, 1] <br> 0 = no linear relationship |
| Geometric EV | $E[X] = 1/p$ | Trials until first success |
| Coupon Collector | $E[T] = n \cdot H_n \approx n \ln n$ | Draws to collect all n types |
| Gambler's Ruin | $P(\text{win}) = k/n$ (fair game) | Start k, target n |
| Sharpe Ratio | $(E[R] - R_f) / \sigma$ | Annualize: multiply by $\sqrt{252}$ |
| Bid-Ask Spread | $P_{\text{ask}} - P_{\text{bid}}$ | Mid = $(P_{\text{ask}} + P_{\text{bid}}) / 2$ |
| Market Impact | $\sigma \sqrt{Q/V}$ | Square-root law (empirical) |
| Put-Call Parity | $C - P = S - Ke^{-rT}$ | Model-free -- always holds |
| Futures Fair Value | $F = S \cdot e^{(r-d)T}$ | r = rate, d = dividend yield |
| Book Imbalance | $(Q_{\text{bid}} - Q_{\text{ask}}) / (Q_{\text{bid}} + Q_{\text{ask}})$ | Predicts short-term price direction |

## Microstructure -- Key Terms

| Term | Definition |
|---|---|
| Bid | Highest price a buyer will pay |
| Ask | Lowest price a seller will accept |
| Spread | Ask minus Bid; market maker's gross revenue per round-trip |
| Mid Price | (Bid + Ask) / 2; theoretical "fair" price |
| Limit Order | Rests on book; provides liquidity; earns rebate |
| Market Order | Executes immediately; takes liquidity; pays fee |
| Adverse Selection | Losing to better-informed counterparties |
| Market Impact | Your own trade moves the price against you |
| Co-location | Servers next to exchange for speed |
| FPGA | Custom hardware; processes data in nanoseconds |
| Tick-to-Trade | Time from data receipt to order sent |
| Book Imbalance | Bid-heavy implies price likely rises; ask-heavy implies falls |
| Spoofing | Fake orders to manipulate book -- illegal |
| Maker-Taker | Provide liquidity = rebate; take liquidity = fee |

## The Greeks -- One-Line Each

| Greek | Symbol | What it measures | One-liner |
|---|---|---|---|
| Delta | $\Delta$ | $\partial V / \partial S$ | "How many shares is this option worth?" |
| Gamma | $\Gamma$ | $\partial^2 V / \partial S^2$ | "How fast does Delta change?" -- highest ATM near expiry |
| Vega | $\mathcal{V}$ | $\partial V / \partial \sigma$ | "How much do I gain or lose per 1% vol move?" |
| Theta | $\Theta$ | $\partial V / \partial t$ | "Time decay: options bleed value daily" |

$V$ = option value, $S$ = underlying price, $\sigma$ = implied volatility, $t$ = time, $r$ = risk-free rate, $\partial$ = partial derivative.

Gamma-Theta tradeoff: $\Theta \sim -(1/2) \Gamma S^2 \sigma^2$. Convexity costs money -- you cannot have Gamma without paying Theta.
