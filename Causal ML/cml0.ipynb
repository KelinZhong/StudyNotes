## Notation key

Used consistently across all notebooks.

| Symbol | Meaning |
|---|---|
| $T$ | Treatment indicator (1 = treated, 0 = control) |
| $Y$ | Observed outcome |
| $Y(1), Y(0)$ | Potential outcomes under treatment / control |
| $X$ | Covariates (confounders / effect modifiers) |
| $Z$ | Instrument (affects T, not Y directly) |
| $e(x) = P(T=1 \mid X=x)$ | Propensity score |
| $m(x) = E[Y \mid X=x]$ | Marginal outcome regression |
| $\mu_t(x) = E[Y \mid T=t, X=x]$ | Arm-specific outcome model |
| $\tau(x) = E[Y(1)-Y(0) \mid X=x]$ | CATE (conditional avg. treatment effect) |
| $\text{ATE} = E[\tau(X)]$ | Average treatment effect |
| $\text{ATT} = E[\tau(X) \mid T=1]$ | Effect on the treated |
| $\text{LATE}$ | Local ATE (compliers only; IV, fuzzy RD) |

## Method selection (canonical cheat sheet)

| Data situation | Method | Identifying assumption |
|---|---|---|
| Can randomize | A/B test (RCT) | Random assignment |
| Observational, confounders measured | PSM · IPW · AIPW | Unconfoundedness + overlap |
| Unmeasured confounders, time variation | DiD | Parallel trends |
| Unmeasured confounders, cutoff rule | RD | Continuity at cutoff |
| Unmeasured confounders, valid instrument | IV / 2SLS | Relevance + exclusion |
| Want individual-level effects | Meta-learners · DML · Causal Forest | Unconfoundedness + overlap |
| Want a targeting policy | Uplift modeling → policy learning | Same + a budget/cost rule |

Rule of thumb: **fewer assumptions is better.** Escalate to an A/B test whenever overlap or parallel trends fails, or the decision is high-stakes.
